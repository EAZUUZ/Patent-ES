```python
import pandas as pd 
import os
import csv
import sys
import os
```

```python
csv.field_size_limit(sys.maxsize)
folder = '/home/eadlzarabi/Desktop/datasets/Patents/'
filename = 'patent_classes.csv'
file = folder + filename
print(file)
```

```python
chunks = pd.read_csv(file, sep=',', quoting=0, lineterminator='\n', dtype=str, chunksize=50000, on_bad_lines='skip')
    
for chunk_idx, chunk in enumerate(chunks):
        print(f"Processing chunk {chunk_idx+1}...")
        chunk.columns = chunk.columns.str.strip()
        print(f"Columns in chunk: {chunk.columns.tolist()}")
        print(f"First row in chunk: {chunk.iloc[0].to_dict()}")
```
